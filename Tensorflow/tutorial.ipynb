<center style="color:white; background-color: blue"><font size = "7">Tensor</center>

Tensor is a generalization of vectors and matrices and is easily understood as a multidimensional array. 

In the general case, an array of numbers arranged on a regular grid with a variable number of axes is known as a tensor.

```python
import tensorflow as tf
```

```python
x = tf.constant([[1, 2, 3], [4, 5, 6]])
```

```python
x
```

```python
print(type(x)) # it tell the type of the tensor
print(x.dtype) # it tell the type of the data in the tensor
```
