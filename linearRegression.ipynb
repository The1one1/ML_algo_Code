### sklearn.datasets has 3 types of datasets:

<span style = "color: orange"> Filling NUll values: </span>:

    df.fillna(value=None, method=None, axis=None, inplace=False, limit=None, downcast=None, **kwargs)

```python
# data = data.dropna() # drop all the rows with null values
df = data.fillna(method='ffill') # fill all the null values with the previous values
    # ffill: forward fill, fill with the previous values
    # bfill: backward fill, fill with the next values 
    # axis=0: fill the values in the same column
    # axis=1: fill the values in the same row
    # inplace=True: modify the original dataframe
```

```python
from sklearn.model_selection import train_test_split
```

```python
# independent features and dependent features

x = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.2, random_state=0)
```

<span style = "color: orange"> Box plot</span>:

            df.boxplot(column=None, by=None, ax=None, fontsize=None, rot=0, grid=True, figsize=None, layout=None, return_type=None, backend=None, **kwargs)
        
    boxplot explaination:
        1. It is used to detect outliers in the data. It is also used to compare the distribution of data between data sets by displaying the 2. median, quartiles, range and outliers. 
        3. In box plot, box represents the 2nd and 3rd quartile (between 25th and 75th percentile).
        4. The line inside the box represents the median (50th percentile).
        5. The whiskers represent the 1st and 4th quartile (between 0th and 25th percentile and 75th and 100th percentile respectively).
        the length of the whiskers is 1.5 times the interquartile range. Interquartile range is the difference between the 3rd quartile and 1st quartile.
        6. The dots outside the whiskers represent the outliers in the data.

<div style="text-align:center"><font size = "5">Scaling</div>

<span style = "color: orange"> StandardScaler</span>:

        StandardScaler is used to standardize the data. It subtracts the mean and divides by the standard deviation, centered around zero. It makes the mean of the distribution 0 and the standard deviation 1.

<span style = "color: orange"> MinMaxScaler/Normalisation</span>:
    
        MinMaxScaler is used to scale the data between 0 and 1. It subtracts the minimum value and divides by the range. It makes the minimum value 0 and the maximum value 1.

<span style = "color: orange"> RobustScaler</span>:
    
        RobustScaler is used to scale the data similar to MinMaxScaler. But it uses the interquartile range instead of the minimum and maximum values. It makes the 1st quartile 0 and the 3rd quartile 1.




<div style="text-align:center"><font size = "5">Note</div>
        
        While applying Scaling, take care of data type. If data type is not float, then it will throw error. So, use try and except block to handle this error.

```python
# standardization
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
```

```python
# apply exception handling
scale.fit(x_train)
```

```python
x_train = scale.transform(x_train)
# x_train = pd.DataFrame(x_train, columns=x.columns).fillna(0)
```

```python
sns.boxplot(x_train)
```

```python
x_test = scale.transform(x_test)
```

```python
sns.boxplot(x_test)
```

```python
# Linear Regression model

from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
pd.DataFrame(x_train).head()
```

```python
model.fit(x_train, y_train)
```

```python
model.coef_
```

```python
model.predict(x_test)
```

```python
# check the accuracy of the model
model.score(x_test, y_test)
```

```python
# also show the predicated line of linear regression
graph = plt.figure(figsize=(6, 6))
plt.scatter(y_test, model.predict(x_test))
# plot y = x line
plt.plot(y_test, y_test, 'r')
plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')
plt.title('Actual Prices vs Predicted Prices')
plt.show()
```
